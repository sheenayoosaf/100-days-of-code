```python
import numpy as np

# Sigmoid activation
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Inputs (column vector)
inputs = np.array([1.0, 0.5])  # shape: (2,)

# Weights — 3 neurons, each with 2 inputs
weights = np.array([
    [0.2, 0.6],
    [0.1, 0.8],
    [0.5, 0.3]
])  # shape: (3, 2)

# Biases — one per neuron
biases = np.array([0.1, -0.2, 0.05])  # shape: (3,)

# Forward pass using matrix multiplication
layer_output = sigmoid(np.dot(weights, inputs) + biases)

print("Layer Output:\n", layer_output)
```
