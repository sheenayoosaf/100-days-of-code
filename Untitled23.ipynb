```python
import cv2                    # OpenCV for image processing
import numpy as np            # Numerical operations
import matplotlib.pyplot as plt  # For plotting images
```

```python
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Make plots a bit bigger
plt.rcParams["figure.figsize"] = (6, 6)

# --------------------------------------------
# 2. Load and display an image with cv2.imread
# --------------------------------------------
# IMPORTANT:
# - Upload an image file (e.g., 'chest_x_ray.jpg') to Colab
#   using the left "Files" panel before running this cell.
# - Or change the filename below to match your image name.

```

```python
image_path = "/content/dog1image.avif"  # Change this if your image has a different name

# cv2.imread reads images in BGR color format by default
image_bgr = cv2.imread(image_path)

if image_bgr is None:
    raise FileNotFoundError(
        f"Could not find image at {image_path}. "
        "Check the file name or upload the image to Colab."
    )

# Convert BGR to RGB for correct display with matplotlib
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.title("Original Image (RGB)")
plt.axis("off")
plt.show()
```

```python
# OpenCV uses BGR, so we convert from BGR to GRAY
image_gray = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)

plt.imshow(image_gray, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")
plt.show()
```

```python
gaussian_blur = cv2.GaussianBlur(image_gray, (5, 5), 0)

plt.imshow(gaussian_blur, cmap="gray")
plt.title("Gaussian Blur")
plt.axis("off")
plt.show()
```

```python
median_blur = cv2.medianBlur(image_gray, 5)

plt.imshow(median_blur, cmap="gray")
plt.title("Median Blur")
plt.axis("off")
plt.show()
```

```python
image_hsv = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2HSV)

# Show RGB
plt.imshow(image_rgb)
plt.title("Image in RGB")
plt.axis("off")
plt.show()
```

```python
plt.imshow(image_hsv)
plt.title("Image in HSV (displayed as RGB)")
plt.axis("off")
plt.show()
```

```python
# Resize image to a fixed size, e.g., 128x128
target_size = (128, 128)
resized_rgb = cv2.resize(image_rgb, target_size)

plt.imshow(resized_rgb)
plt.title("Resized Image (128x128)")
plt.axis("off")
plt.show()
```

```python
rotated_90 = cv2.rotate(resized_rgb, cv2.ROTATE_90_CLOCKWISE)

plt.imshow(rotated_90)
plt.title("Rotated 90 Clockwise")
plt.axis("off")
plt.show()



```

```python
# Flip image horizontally (left-right)
flipped_horizontal = cv2.flip(resized_rgb, 1)

plt.imshow(flipped_horizontal)
plt.title("Flipped Horizontally")
plt.axis("off")
plt.show()
```

```python
# Flip image vertically (up-down)
flipped_vertical = cv2.flip(resized_rgb, 0)

plt.imshow(flipped_vertical)
plt.title("Flipped Vertically")
plt.axis("off")
plt.show()
```

```python
sharpen_kernel = np.array([
    [0, -1,  0],
    [-1, 5, -1],
    [0, -1,  0]
], dtype=np.float32)

# Apply to grayscale resized image for simplicity
resized_gray = cv2.cvtColor(resized_rgb, cv2.COLOR_RGB2GRAY)
sharpened = cv2.filter2D(resized_gray, -1, sharpen_kernel)

plt.imshow(sharpened, cmap="gray")
plt.title("Sharpened Image (Custom Kernel)")
plt.axis("off")
plt.show()
```

```python
input_shape = (128, 128, 1)  # height, width, channels

simple_model = Sequential([
    Conv2D(
        filters=16,
        kernel_size=(3, 3),
        activation="relu",
        input_shape=input_shape,
        name="conv_1"
    )
])
```

```python
input_shape = (128, 128, 1)  # height, width, channels

simple_model = Sequential([
    Conv2D(
        filters=16,
        kernel_size=(3, 3),
        activation="relu",
        input_shape=input_shape,
        name="conv_1"
    )
])

# Show the model architecture
simple_model.summary()

# --------------------------------------------
# 9. Prepare one image as input to the CNN
# --------------------------------------------
# We already have resized_gray (128x128). Let's normalize it and add batch/channel dims.

# Normalize pixel values to [0, 1]
img_norm = resized_gray.astype("float32") / 255.0

# Add channel dimension: (128, 128) -> (128, 128, 1)
img_norm = np.expand_dims(img_norm, axis=-1)

# Add batch dimension: (128, 128, 1) -> (1, 128, 128, 1)
img_batch = np.expand_dims(img_norm, axis=0)

print("Input batch shape for model:", img_batch.shape)

# --------------------------------------------
# 10. Get feature maps from the single Conv2D layer
# --------------------------------------------
# We create a new model that takes the same input but outputs the activations
# of the first convolution layer.

feature_extractor = Model(
    inputs=simple_model.inputs,
    outputs=simple_model.layers[0].output
)

feature_maps = feature_extractor.predict(img_batch)

print("Feature maps shape:", feature_maps.shape)
# Expected shape: (1, new_height, new_width, 16)

# Remove batch dimension for plotting: (h, w, 16)
feature_maps = feature_maps[0]

# --------------------------------------------
# 11. Visualize feature maps as images
# --------------------------------------------
num_filters = feature_maps.shape[-1]

# Decide how many rows and cols to show (e.g., 4x4 for 16 filters)
rows = 4
cols = 4

plt.figure(figsize=(8, 8))
for i in range(num_filters):
    plt.subplot(rows, cols, i + 1)
    feature_map = feature_maps[:, :, i]
    plt.imshow(feature_map, cmap="gray")
    plt.axis("off")
    plt.title(f"F{i+1}")
plt.suptitle("Feature Maps from First Conv2D Layer", y=0.92, fontsize=14)
plt.tight_layout()
plt.show()
```

```python
extended_model = Sequential([
    Conv2D(16, (3, 3), activation="relu", input_shape=input_shape, name="conv1"),
    MaxPooling2D((2, 2), name="pool1"),

    Conv2D(32, (3, 3), activation="relu", name="conv2"),
    MaxPooling2D((2, 2), name="pool2"),

    Flatten(name="flatten"),
    Dense(64, activation="relu", name="dense1"),
    Dense(1, activation="sigmoid", name="output")  # Example: binary classifier
])

print("\nExtended CNN Architecture:")
extended_model.summary()
```
