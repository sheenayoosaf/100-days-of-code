```python
import numpy as np

# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Example input (2 features)
inputs = np.array([1.0, 0.5])  # shape = (2,)

# Hidden layer weights and biases
weights_hidden = np.array([
    [0.2, 0.6],
    [0.1, 0.8],
    [0.5, 0.3]
])  # shape = (3,2)
bias_hidden = np.array([0.1, -0.2, 0.05])  # shape = (3,)

# Output layer weights and bias
weights_output = np.array([[0.4, -0.3, 0.2]])  # shape = (1,3)
bias_output = np.array([0.1])  # shape = (1,)

# --- Forward Pass ---

# Hidden layer output
hidden_layer_input = np.dot(weights_hidden, inputs) + bias_hidden
hidden_layer_output = sigmoid(hidden_layer_input)

# Output layer
final_input = np.dot(weights_output, hidden_layer_output) + bias_output
final_output = sigmoid(final_input)

print("Hidden Layer Output:", hidden_layer_output)
print("Final Output:", final_output)
```

We created a tiny brain that makes a prediction.

It has:

Input Layer → Hidden Layer → Output Layer
     2             3             1
     Neuron Output = sigmoid(input1*w1 + input2*w2 + ... + bias)
